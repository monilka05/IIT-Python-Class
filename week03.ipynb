### if else

```python
num=int(input("Enter number."))
if num>0:
    print("Positive number")
elif num<0:
    print("negative number")
else:
    print("Zero")
```

### nested if else

```python
num=int(input("Enter number."))
if num>0:
    print("Positive number")
else:
    if num<0:
        print("Negative number")
    else: print("0")
```
